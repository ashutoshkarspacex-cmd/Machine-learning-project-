```python
df['label']=df['label'].map({1:'spam',0:'not spam'})
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['label']=le.fit_transform(df['label'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
corpus=df['email']
x=vectorizer.fit_transform(corpus)
```

```python
df
```

```python
y=df['label']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import *
accuracy_score(y_pred,y_test)*100
```

```python
f1_score(y_pred,y_test)*100
```

```python
precision_score(y_pred,y_test)*100
```

```python
k=classification_report(y_pred,y_test)
print(k)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
roc_auc_score(y_test,y_pred).astype(float)
```

```python
svc_disp = RocCurveDisplay.from_estimator(lr, x_test, y_test)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Spam", "Not Spam"])
disp.plot(cmap=plt.cm.Blues)
```
